```python
# 取出更新後的資料
def get_update_data(df_title):
    # 拿之前的檔案
    try:
        df_history = pd.read_excel('history.xlsx')
        df_history['old'] = 'Y'
        df_history = df_history[['scode', 'date', 'time', 'old']]
        df_history = df_history.astype(str)

        # 把之前的檔案跟新爬到的資料作比對
        df_new = df_title.merge(df_history, on=['scode', 'date', 'time'] , how='left')
        df_new = df_new[df_new['old'].isnull()]
    except:
        df_new = df_title.copy()

    # 更新之前的檔案
    # df_title.to_excel('history.xlsx', index=False)

    return df_new

df_new = get_update_data(df_title)
df_new
```
